```python
instruments = df['Instrument'].unique()
```

```python
# Split each unique instrument string by semicolon and flatten the result into a list
import numpy as np
instruments_split = []
for instrument in instruments:
    if pd.isnull(instrument):
        continue
    instruments_split.extend([item.strip() for item in str(instrument).split(';')])
instruments_split = np.array(instruments_split)
len(instruments_split)
```

```python
#remove the duplicates from instruments split
instruments_split = np.unique(instruments_split)
len(instruments_split)
```

```python
instruments_split
```

```python
potential_instruments = [
    'Climate finance tools|Economic',
    'Provision of climate funds|Direct Investment',
    'Carbon pricing & emissions trading|Economic',
    'Subsidies|Economic',
    'Tax incentives|Economic',
    'Green procurement|Direct Investment',
    'Insurance|Economic',
    'Other|Economic', 
    'Other|Direct Investment',
    'Disclosure obligations|Regulation',
]

optional_instruments = ['Education, training and knowledge dissemination|Information',
       'Institutional mandates|Governance', 'Insurance|Economic',
       'Moratoria & bans|Regulation',
       'Research & Development, knowledge generation|Information',
       'Standards, obligations and norms|Regulation']
```

```python
#check if there are any rows where the instrument column is null or missing 
missing_instruments = df[df['Instrument'].isnull()]
missing_instruments
#check if there are any rows where the instrument column is empty
empty_instruments = df[df['Instrument'] == '']
empty_instruments

```

```python
# Find all rows where Instrument contains any of the instruments in potential_instruments or optional_instruments
all_instruments = potential_instruments + optional_instruments
finance_df = df[df['Instrument'].apply(lambda x: any(instr in str(x) for instr in all_instruments))]

# Get counts for each potential and optional instrument
instrument_counts = {instr: finance_df['Instrument'].apply(lambda x: instr in str(x)).sum() for instr in all_instruments}
print("Number of documents containing each instrument (potential and optional):")
for instr, count in instrument_counts.items():
    print(f"{instr}: {count}")
```

```python
excluded_document_types = [
    'Press Release',
    'Summary',
    'Statement',
    'Vision',
    'Agenda',
    'Annex',
    'Progress Report',
    'Strategic Assessment',
    'Report',
    'Publication',
    'Criteria',
    'Discussion Paper'
]

filtered_df = finance_df[~finance_df['Document Type'].str.strip().str.casefold().isin([x.casefold() for x in excluded_document_types])]
```
